```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("Mr. Anubhav Gupta loves coding and soving problems. He wanted to become data scientist.")

for sentence in doc.sents:
    print(sentence)
```

```python
for sentence in doc.sents:
    for word in sentence:
        print(word)
```

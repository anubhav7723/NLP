```python
nlp = spacy.load("en_core_web_sm")

doc = nlp("We just opened our wings, the flying part is comming soon")

for token in doc:
    if token.is_stop:
        print(token)
```

```python
def preprocess(text):
    doc = nlp(text)
    
    not_stop_words =[token.text for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(not_stop_words)
```

```python
preprocess("We just opened our wings, the flying part is comming soon")
```

```python
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from sklearn.feature_extraction.text import CountVectorizer
```

```python

v = CountVectorizer(ngram_range=(1,2))

v.fit(["Anubhav is looking for a job"])
v.vocabulary_
```

```python
corpus = [
    "Anubhav ate apple",
    "Soham plays cricket",
    "Rahul is playing video game"
]
```

```python
nlp = spacy.load("en_core_web_sm")


def preprocess(text):
    doc = nlp(text)
    filtered_tokens = []
    for token in doc:
        if not token.is_stop and not token.is_punct:
            filtered_tokens.append(token.lemma_)
    return " ".join(filtered_tokens)
```

```python
corpus_new = [preprocess(text) for text in corpus]
corpus_new
```

```python
v = CountVectorizer(ngram_range=(1,2))

v.fit(corpus_new)
v.vocabulary_
```

```python
v.transform(["Anubhav playing video game"]).toarray()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json("news_dataset.json")
df.head()
```

```python
df.shape
```

```python
df['category'].value_counts()
```

```python
min_samples = 1381

df_business = df[df.category=='BUSINESS'].sample(min_samples,random_state=10)
df_sports = df[df.category=='SPORTS'].sample(min_samples,random_state=10)
df_crime = df[df.category=='CRIME'].sample(min_samples,random_state=10)
df_science = df[df.category=='SCIENCE'].sample(min_samples,random_state=10)
```

```python
df_balanced = pd.concat([df_business,df_sports,df_crime,df_science],axis=0)
df_balanced
```

```python
target = {'BUSINESS': 0, 'SPORTS': 1, 'CRIME': 2, 'SCIENCE': 3}

df_balanced['category_num'] = df_balanced['category'].map({
    'BUSINESS': 0,
    'SPORTS': 1, 
    'CRIME': 2, 
    'SCIENCE': 3
})
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    df_balanced.text, 
    df_balanced.category_num, 
    test_size=0.2,
    random_state=10,
    stratify=df_balanced.category_num
)
```

```python
print(X_train.shape)
X_train.head()
```

```python
y_train.value_counts()
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report

clf = Pipeline([
     ('vectorizer_bow', CountVectorizer(ngram_range = (1, 1))),    
     ('Multi NB', MultinomialNB())         
])

clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
#1. create a pipeline object
clf = Pipeline([
    ('vectorizer_1_2_gram', CountVectorizer(ngram_range = (1, 2))),        #using the ngram_range parameter 
     ('Multi NB', MultinomialNB())         
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

```python
df_balanced['preprocessed_txt'] = df_balanced['text'].apply(preprocess) 
```

```python
df_balanced.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df_balanced.preprocessed_txt, 
    df_balanced.category_num, 
    test_size=0.2, # 20% samples will go to test dataset
    random_state=2022,
    stratify=df_balanced.category_num
)
```

```python
print(X_train.shape)
X_train.head()
```

```python
#1. create a pipeline object
clf = Pipeline([
    ('vectorizer_bow', CountVectorizer(ngram_range = (1, 2))),        #using the ngram_range parameter 
    ('Multi NB', MultinomialNB())
])

#2. fit with X_train and y_train
clf.fit(X_train, y_train)


#3. get the predictions for X_test and store it in y_pred
y_pred = clf.predict(X_test)


#4. print the classfication report
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from matplotlib import pyplot as plt
import seaborn as sn
plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Prediction')
plt.ylabel('Truth')
```

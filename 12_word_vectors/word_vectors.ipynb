```python
base_token = nlp('banana')
base_token.vector.shape
```

```python
doc = nlp('i ate some fruits like banana , apple , orange fake word car truck')

for token in doc:
    print(f"{token.text} <-> {base_token}",token.similarity(base_token))
```

```python
def printSimilarity(base_word , text):
    base_token = nlp(base_word)
    doc = nlp(text)
    for token in doc:
        print(f"{token.text} <-> {base_token}",token.similarity(base_token))

printSimilarity('iphone' , 'tesla samsung truck porsche phone banana')
```

```python
king = nlp.vocab["king"].vector
man = nlp.vocab["man"].vector
woman = nlp.vocab["woman"].vector
queen = nlp.vocab["queen"].vector

result = king - man + woman
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_similarity([result], [queen])
```
